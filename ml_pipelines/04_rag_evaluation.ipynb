# RAG Retrieval Benchmarking\n**Objective:** Compute Recall@K, Precision@K, and Mean Reciprocal Rank (MRR) for our pgvector/hybrid search parameters.\n---\nOptimized for NVIDIA GeForce RTX 4080 (16GB VRAM) leveraging CUDA, PyTorch, and PEFT libraries.

```python
import numpy as np

# Evaluate retrieval precision
retrieved = ["~세요 usage rules", "~네요 exclamation rules", "Hangeul basics"]
ground_truth = "~네요 exclamation rules"

mrr = 0.0
for idx, doc in enumerate(retrieved):
    if doc == ground_truth:
        mrr = 1.0 / (idx + 1)
        break

print(f"Evaluated Mean Reciprocal Rank (MRR): {mrr}")
```
