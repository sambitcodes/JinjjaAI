# High-Performance Local Inference\n**Objective:** Accelerate local inference of fine-tuned models utilizing quantized bitsandbytes or HuggingFace transformers.\n---\nOptimized for NVIDIA GeForce RTX 4080 (16GB VRAM) leveraging CUDA, PyTorch, and PEFT libraries.

```python
import torch
from transformers import pipeline

generator = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B-Instruct", device=0)
prompt = "Explain '~고 싶다' grammar pattern in Korean."
output = generator(prompt, max_length=150)
print(output[0]["generated_text"])
```
