# Local Embedding Indexing using BGE-M3\n**Objective:** Index lesson markdown documents to local vector coordinates utilizing CUDA-accelerated BGE-M3 embeddings.\n---\nOptimized for NVIDIA GeForce RTX 4080 (16GB VRAM) leveraging CUDA, PyTorch, and PEFT libraries.

```python
import torch
from sentence_transformers import SentenceTransformer

# Verify RTX 4080 CUDA acceleration
print("CUDA Available:", torch.cuda.is_available())
device = "cuda" if torch.cuda.is_available() else "cpu"

model = SentenceTransformer("BAAI/bge-m3", device=device)
sentences = ["안녕하세요", "한국어 공부는 재미있습니다."]
embeddings = model.encode(sentences, normalize_embeddings=True)
print("Embedding dimension:", embeddings.shape[1])
```
