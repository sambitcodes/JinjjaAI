# Fine-Tuned Model Benchmarking\n**Objective:** Benchmark perplexity, BLEU, and Korean grammatical accuracy against baseline parameters.\n---\nOptimized for NVIDIA GeForce RTX 4080 (16GB VRAM) leveraging CUDA, PyTorch, and PEFT libraries.

```python
from nltk.translate.bleu_score import sentence_bleu

reference = [["안녕하세요", "저는", "학생입니다"]]
candidate = ["안녕하세요", "저는", "학생입니다"]
score = sentence_bleu(reference, candidate)
print(f"BLEU evaluation score: {score * 100:.2f}%")
```
