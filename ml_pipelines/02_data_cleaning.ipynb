# Data Cleaning & De-duplication\n**Objective:** Deduplicate datasets, enforce correct Korean Unicode normalization (NFC), and filter low-quality syntax patterns.\n---\nOptimized for NVIDIA GeForce RTX 4080 (16GB VRAM) leveraging CUDA, PyTorch, and PEFT libraries.

```python
import unicodedata
import json

def clean_and_normalize(text):
    # Enforce NFC Unicode normalization for Korean syllables
    return unicodedata.normalize('NFC', text.strip())

with open("tutor_sft_dataset.json", "r", encoding="utf-8") as f:
    data = json.load(f)

cleaned = []
seen = set()
for item in data:
    instr = clean_and_normalize(item["instruction"])
    out = clean_and_normalize(item["output"])
    if instr not in seen:
        seen.add(instr)
        cleaned.append({"instruction": instr, "output": out})

print(f"Original size: {len(data)}, Cleaned size: {len(cleaned)}")
```
